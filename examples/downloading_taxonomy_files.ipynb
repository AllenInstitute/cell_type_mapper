# Downloading the files that define our taxonomies

As of autumn 2025, the files that define the on-line MapMyCells taxonomies are available through [abc_atlas_access](https://alleninstitute.github.io/abc_atlas_access/intro.html), the Python API for downloading data from ABC Atlas. In this notebook, we will show what files are available through `abc_atlas_access` and how to download them onto your machine.

`abc_atlas_access` manages data through a cache object. The cache object is associated with a specific directory in your system. When you ask for a file from `abc_atlas_access`, it checks to see if that file has already been downloaded. If it has not, it downloads that file. Regardless, the cache then returns the path to the file on your local system (for a more detailed explanation, see the official `abc_atlas_access` ["getting started" notebook](https://alleninstitute.github.io/abc_atlas_access/notebooks/getting_started.html)).

# Setting up abc_atlas_access

First, let's create a cache object to download the data files we are discussing.

**Note:** as of version `1.6.0`, `abc_atlas_access` is a dependency of `cell_type_mapper`, so modern installations of `cell_type_mapper` should just have `abc_atlas_access`.

```python
import abc_atlas_access.abc_atlas_cache.abc_project_cache as cache_module

import os
import pathlib
```

```python
cache_dir = "/Volumes/mazarbul/workspace/abc_cache"

if not pathlib.Path(cache_dir).is_dir():
    raise RuntimeError(
        f"\n{cache_dir}\nis not a valid directory on your system\n"
        "set cache_dir above to a directory where you wish to download "
        "and store data from abc_atlas_access."
    )
```

```python
abc_cache = cache_module.AbcProjectCache.from_cache_dir(cache_dir)
```

# Available taxonomies

Datasets in `abc_atlas_access` are separated into different "directories". You can see the directories available through `abc_atlas_access` with

```python
abc_cache.list_directories
```

Within each directory are various files that `abc_atlas_access` provides access to. To list the files related to MapMyCells, you can run

```python
abc_cache.list_mapmycells_files(directory="WMB-taxonomy")
```

Asking for the MapMyCells files in a directory that does not contain any MapMyCells files will result in an error.

```python
abc_cache.list_mapmycells_files(directory="WMB-10X")
```

The directories that contain MapMyCells files are:

```python
for directory_name in abc_cache.list_directories:
    try:
        file_list = abc_cache.list_mapmycells_files(directory=directory_name)
        print(f"{directory_name} : {file_list}")
    except Exception:
        pass
```

# Downloading Whole Mouse Brain taxonomy MapMyCells files

To download the `precomputed_stats` file and the marker gene lookup table associated with the Whole Mouse Brain taxonomy on MapMyCells, run

```python
def get_file_size(file_path):
    """
    Take the path to a file. Print its size in GB to stdout
    """
    gb = os.stat(file_path).st_size/(1024**3)
    print(f"{file_path.name} is {gb:.2e} GB in size")
```

```python
abc_cache.list_mapmycells_files(directory="WMB-taxonomy")
```

```python
wmb_marker_path = abc_cache.get_file_path(
    directory="WMB-taxonomy",
    file_name="mouse_markers_230821"
)
```

```python
wmb_marker_path
```

```python
get_file_size(wmb_marker_path)
```

```python
wmb_precomputed_path = abc_cache.get_file_path(
    directory="WMB-taxonomy",
    file_name="precomputed_stats_ABC_revision_230821"
)
```

```python
wmb_precomputed_path
```

```python
get_file_size(wmb_precomputed_path)
```

The files now exist on your system in the locations listed above.
